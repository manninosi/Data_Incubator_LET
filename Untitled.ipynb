```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


file_01= "county_lex_2020-01-20.csv.gz"
file_02= "county_lex_2020-02-01.csv.gz"
file_03= "county_lex_2020-03-01.csv.gz"
file_04= "county_lex_2020-04-14.csv.gz"

df_01 	= pd.read_csv(file_01, compression='gzip', header=0)
df_02 	= pd.read_csv(file_02, compression='gzip', header=0)
df_03 	= pd.read_csv(file_03, compression='gzip', header=0)
df_04 	= pd.read_csv(file_04, compression='gzip', header=0)
```

```python
#Selecting mississppi counties 
#Jan
df_01_MS = pd.concat([df_01.loc[:,"COUNTY_PRE"],df_01.loc[:,"28001":"28163"]], axis=1)
df_01_MS = df_01_MS[df_01_MS['COUNTY_PRE']>=28001]
df_01_MS = df_01_MS[df_01_MS['COUNTY_PRE']<=28163]
#Feb
df_02_MS = pd.concat([df_02.loc[:,"COUNTY_PRE"],df_02.loc[:,"28001":"28163"]], axis=1)
df_02_MS = df_02_MS[df_02_MS['COUNTY_PRE']>=28001]
df_02_MS = df_02_MS[df_02_MS['COUNTY_PRE']<=28163]
#Mar
df_03_MS = pd.concat([df_03.loc[:,"COUNTY_PRE"],df_03.loc[:,"28001":"28163"]], axis=1)
df_03_MS = df_03_MS[df_03_MS['COUNTY_PRE']>=28001]
df_03_MS = df_03_MS[df_03_MS['COUNTY_PRE']<=28163]
#Apr
df_04_MS = pd.concat([df_04.loc[:,"COUNTY_PRE"],df_04.loc[:,"28001":"28163"]], axis=1)
df_04_MS = df_04_MS[df_04_MS['COUNTY_PRE']>=28001]
df_04_MS = df_04_MS[df_04_MS['COUNTY_PRE']<=28163]
frame_MS = {'Jan': df_01_MS.iloc[:,1:].mean(), 'Feb':df_02_MS.iloc[:,1:].mean(), 'Mar':df_03_MS.iloc[:,1:].mean(), 'Apr':df_04_MS.iloc[:,1:].mean()}

MS_LET_mean = pd.DataFrame(frame_MS)
ax = MS_LET_mean.plot(kind='box',  fontsize =14)
ax.set_xlabel('Month',size = 16)
ax.set_ylabel('LET', size = 16)
ax.set_title("Mississppi LET values for All Counties", size = 18)
plt.savefig('MS_LET.eps', formtat='eps')
```
